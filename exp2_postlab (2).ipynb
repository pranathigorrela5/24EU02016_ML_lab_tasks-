```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

df = sns.load_dataset("titanic")

print("Original Dataset")
print(df.head())

print("\nMissing Values Before Preprocessing")
print(df.isnull().sum())

df.isnull().sum().plot(kind="bar")
plt.title("24EU02016_Missing Values")
plt.xlabel("Columns")
plt.ylabel("Missing Count")
plt.show()

df["age"] = df["age"].fillna(df["age"].median())
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])

df.drop(columns=["deck", "embark_town", "alive", "class", "who"], inplace=True)

encoder = LabelEncoder()

for col in ["sex", "adult_male", "alone", "embarked"]:
    df[col] = encoder.fit_transform(df[col].astype(str))

print("\nMissing Values After Preprocessing")
print(df.isnull().sum())

print("\nPreprocessed Dataset")
print(df.head())

df["survived"].value_counts().plot(kind="bar")
plt.title("24EU02016_Survival Distribution")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()

df["age"].plot(kind="hist")
plt.title("24EU02016_Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
